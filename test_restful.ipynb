###  urllib.request访问带参数的POST请求

```python
import urllib.request
import urllib.parse

import requests
import base64
```

```python
host  = 'http://192.168.30.110:5000/recognize_vege'

requestData = {}

test_path = './test_pics/qingcai-03.jpg'

with open(test_path, 'rb') as f:
    picData = f.read()
    bb64 = base64.b64encode(picData)
    requestData['pic_str'] = bb64
    
request = urllib.request.Request(host)
    
request.add_header("Content-Type","application/x-www-form-urlencoded;charset=utf-8")
f = urllib.request.urlopen(request, requestData)

print (f.read())

```
